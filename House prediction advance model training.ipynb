```python
import pandas as pd
import numpy as np  

pd.pandas.set_option('display.max_columns',None)
```

```python
df = pd.read_csv(r'/content/drive/MyDrive/CSV/X_train.csv')
```

```python
df.head()
```

```python
Y_train = df['SalePrice']
```

```python
X_train = df.drop(['SalePrice','Id'],axis = 1)
```

```python
X_train.head()
```

```python
X_train.shape
```

```python
Y_train.shape
```

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

#it selects which features are imp to consider and which are not out of all
#value of alpha defines what amount of column to take ......it should be between 0.001 - 5 only
selected_features = SelectFromModel(Lasso(alpha = 0.005,random_state = 0))
selected_features.fit(X_train,Y_train)
```

```python
selected_features.get_support()
```

```python
# let's print the number of total and selected features

# this is how we can make a list of the selected features
selected_feat = X_train.columns[(selected_features.get_support())]

# let's print some stats
print('total features: {}'.format((X_train.shape[1])))
print('selected features: {}'.format(len(selected_feat)))
print('features with coefficients shrank to zero: {}'.format(
    np.sum(selected_features.estimator_.coef_ == 0)))
```

```python
X_train[selected_feat].head()
```

```python
df_final = X_train[selected_feat].copy()
```

```python
df_final.head()
```

```python
df_final = pd.concat([df_final,Y_train],axis = 1)
```

```python
df_final.head()
```

```python
y = df_final['SalePrice']
```

```python
X = df_final.drop('SalePrice',axis = 1)
```

```python
print("X shape :",X.shape)
print("y shape :",y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
ann = Sequential()
ann.add(Dense(units = 6, activation = 'relu', input_dim = 21))
ann.add(Dense(units = 6, activation = 'relu'))
ann.add(Dense(units = 2, activation = 'sigmoid'))
ann.compile(optimizer = 'adam', loss = 'mean_squared_error', metrics = ['accuracy'])
ann.fit(X_train, y_train, batch_size = 10, epochs = 100)
```
